# Import Libraries


```python
#data analysis
import pandas as pd
#data visualization
import matplotlib.pyplot as plt
import plotly.express as px

#download files to our computer
from google.colab import files
```

#Import Data
download data from this github [repository](https://github.com/cmclane1/Covid-in-Americas-Largest-Cities)


```python
#import csv of covid vs income data

#save github csv link
covid_data = "https://raw.githubusercontent.com/cmclane1/Covid-in-Americas-Largest-Cities/main/Covid_data.csv"
#define our initial data frame
df_covid = (pd.read_csv(covid_data, thousands=","))
df_covid.drop([0])


```

#Look at data


```python
df_covid = df_covid.head(47)
df_covid
```

```python
#general stats to understand data
new_df = df_covid[['Population', 'Household Income', 'Covid cases', 'Incident Rate', 'Case_Fatality_Ratio']]
new_df.apply(pd.to_numeric)
new_df.describe(include = 'all')



```

#Data Visualizations with Plotly
Covid Cases as Y

```python
#make line graph in plotly express
covid_cases_as_y = px.scatter(data_frame=df_covid, x = "Population", y = "Covid cases", title = "Population vs Cases", trendline="ols")
covid_cases_as_y
```

continued

```python
covid_cases_as_y2 = px.scatter(data_frame=df_covid, x = "Household Income", y = "Covid cases", title = "Household Income vs Cases", trendline="ols")
covid_cases_as_y2
```
